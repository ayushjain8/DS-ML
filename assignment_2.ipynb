```python
base_fare = float(input("Enter base ticket fare: ₹"))
age = int(input("Enter passenger age: "))

if age <= 5:
    fare = 0
    message = "Ticket is FREE for children under 5."
elif age <= 12:
    fare = base_fare * 0.5
    message = "50% discount applied for child fare."
elif age >= 60:
    fare = base_fare * 0.7
    message = "30% discount applied for senior citizen."
else:
    fare = base_fare  # No discount
    message = "Full fare applicable."


print(message)
print(f"Total payable fare: ₹{fare:.2f}")
```

```python
units = float(input("Enter electricity units consumed: "))

if units <= 50:
    amount = units * 3.5
elif units <= 150:
    amount = (50 * 3.5) + ((units - 50) * 4.0)
elif units <= 250:
    amount = (50 * 3.5) + (100 * 4.0) + ((units - 150) * 5.2)
else:
    amount = (50 * 3.5) + (100 * 4.0) + (100 * 5.2) + ((units - 250) * 6.5)

print(f"Total electricity bill for {units} units: ₹{amount:.2f}")
```

```python
days_late = int(input("Enter the number of overdue days: "))

if days_late <= 0:
    fine = 0
    message = "No fine. Book returned on time!"
elif days_late <= 5:
    fine = days_late * 2  # ₹2 per day for first 5 days
    message = "₹2/day fine for first 5 days."
elif days_late <= 10:
    fine = (5 * 2) + (days_late - 5) * 5  # ₹5/day for next 5 days
    message = "₹2/day for first 5 days, ₹5/day for next 5."
else:
    fine = (5 * 2) + (5 * 5) + (days_late - 10) * 10  # ₹10/day after 10 days
    message = "₹2/day for 5 days, ₹5/day for next 5, ₹10/day afterwards."

print(message)
print(f"Total library fine: ₹{fine}")

```
