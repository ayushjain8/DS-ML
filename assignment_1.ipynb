```python

marks = int(input("Enter your marks (0–100): "))


if marks >= 90:
    grade = "A+"
elif marks >= 80:
    grade = "A"
elif marks >= 70:
    grade = "B"
elif marks >= 60:
    grade = "C"
elif marks >= 50:
    grade = "D"
elif marks >= 40:
    grade = "E"
else:
    grade = "F (Fail)"


print("Your grade is:", grade)
```

```python

num1 = float(input("Enter first number: "))
operation = input("Choose operation (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Cannot divide by zero"
else:
    result = "Invalid operation"


print("Result:", result)
```

```python

age = int(input("Enter your age: "))

if age >= 21:
    print("You are eligible for Voting, Driving, and Drinking.")
elif age >= 18:
    print("You are eligible for Voting and Driving, but NOT for Drinking.")
else:
    print("You are NOT eligible for Voting, Driving, or Drinking.")
```
