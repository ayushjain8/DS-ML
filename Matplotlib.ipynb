```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])  # X-axis
y = np.array([1, 2, 3, 4])  # Y-axis

plt.plot(x, y)
plt.show()
```

```python
#BAR GRAPH


import matplotlib.pyplot as plt
import numpy as np

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

plt.bar(fruits, sales)
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(1000)
plt.hist(data ,bins=30, color='hotpink')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=[32,56,74,68,90]
s=[41,68,62,91,31]
plt.scatter(a,s)
plt.xlabel('values')
plt.ylabel('numbers')
plt.title('Scatter Plot')
plt.show()
```

```python
fruits = ['Papaya', 'Bananas', 'Apple', 'Dates', 'Kiwi']
data = [10,40,77,42,89]
plt.pie(data, labels=fruits)
plt.title('Fruit Sales')
plt.show()
```
